### Transformação

<ul><li>O dataset em questão possui apenas uma coluna com valores numéricos e esse fato resultou em várias trasnformações;</li>
    <li>As colunas reviews, size, installs e price deveriam ser do tipo numérico;</li>
    <li>Depois de um tempo trabalhando com ele concluí que não foi uma boa escolha;</li>
</ul>

```python
#Função que verifica quantidade de nulos por coluna
def containsNull(x):
  return sum(x.isnull())
```

```python
playStore.apply(containsNull, axis=0)
```

```python
#Preenchendo todos os valores null das colunas apenas em memória
playStore.fillna(0,inplace=True)
```

```python
playStore.apply(containsNull, axis=0)
```

```python
#Para converter os campos foi necessário tratar alguns dados incorretos do dataset
playStore['Price'] = playStore['Price'].apply(lambda x: str(x).replace('$','')) 
```

```python
playStore['Price'] = playStore['Price'].apply(lambda x: str(x).replace('Everyone','0.0')) 
```

```python
playStore['Type'] = playStore['Type'].apply(lambda x: str(x).replace('0','Free')) 
```

```python
playStore['Type'] = playStore['Type'].apply(lambda x: str(x).replace('nan','Free')) 
```

```python
playStore['Reviews'] = playStore['Reviews'].apply(lambda x: str(x).replace('3.0M','100')) 
```

```python
playStore['Installs'] = playStore['Installs'].apply(lambda x: str(x).replace('Free','10,000+')) 
```

```python
#Conversão do campo object para numeric
pd.to_numeric(playStore['Price'])
```

```python
#Conversão do campo object para numeric
pd.to_numeric(playStore['Reviews'])
```

### Visualização

```python
#Gráfico contendo os 5 gêneros mais baixados na plataforma do google
playStore['Genres'].value_counts()
```

```python
labels = 'Tools','Entertainment', 'Education', 'Medical', 'Business', 
sizes = [842, 623, 549, 463, 460]

colors = ['#C39BD3', # Tools
          '#A9CCE3',  # Entertainment
          '#F9E79F',  # Education
          '#7DCEA0',  # Medical
          '#F08080',  # Business               
          ]

explode = (0, 0, 0, 0, 0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', startangle=60)
plt.axis('equal')
plt.title("% Top 5 Genres", fontsize=15)
plt.plot()
fig=plt.gcf()
fig.set_size_inches(10,10)
plt.show()
```

```python
#Categorias
playStore['Category'].value_counts()
```

```python
#Gráfico representando as categorias e tipos dos apps
plt.figure(figsize = (20,8))
ax = sns.countplot(x="Category", hue="Type", order=playStore['Category'].value_counts().index, data=playStore, palette="Paired")
plt.xticks(rotation= 90)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(['FREE', 'PAID'], loc=9, fontsize=15)
plt.xlabel('Category', fontsize=15)
plt.ylabel('Total', fontsize=15)
plt.title('Category games and your types',fontsize=21)
```
